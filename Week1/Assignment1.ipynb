```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("sample.jpg")
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python


def process_meme(image_path):
    img = Image.open(image_path)
    width, height = img.size
    print(f"Original Dimensions: Width={width}, Height={height}")
    gray_img = img.convert('L')
    resized_img = gray_img.resize((width // 2, height // 2))
    resized_img.save("processed_week1.jpg")
    print("Processed image saved as 'processed_week1.jpg'")
```

```python
process_meme("sample.jpg")
```
